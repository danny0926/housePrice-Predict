```python
def rmse_cv(model,X,y):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=5))
    return rmse

def algorithm_pipeline(train_x, train_y, model, param_grid, cv=5, scoring_fit='neg_mean_squared_error'):
     
    gs = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=cv,
        n_jobs=-1,
        scoring=scoring_fit,
        verbose=2
    )
     
    fitted_model = gs.fit(train_x, train_y)
    best_model = fitted_model.best_estimator_
    best = fitted_model.best_params_
    fitted_model.cv_results_['mean_test_score']= np.sqrt(-fitted_model.cv_results_['mean_test_score'])
    data_score = pd.DataFrame( fitted_model.cv_results_)[['params','mean_test_score','std_test_score']]
     
    return [best_model,best,data_score]

def get_mape(y_true, y_pred):
    """
    Compute mean absolute percentage error (MAPE)
    """
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def _load_data(data_path, mode='train'):
    df = pd.read_csv(data_path, sep=",")    
    df = df.drop(['ID', '備註'], axis=1)

#     city = df["縣市"].unique()
#     town = df["鄉鎮市區"].unique()
#     road = df["路名"].unique()
#     district = df["使用分區"].unique()
#     use = df["主要用途"].unique()
#     materials = df["主要建材"].unique()
#     structtype = df["建物型態"].unique()
    
#     city_map = {x : i for i, x in enumerate(city)}
#     town_map = {x : i for i, x in enumerate(town)}
#     road_map = {x : i for i, x in enumerate(road)}
#     district_map = {x : i for i, x in enumerate(district)}
#     use_map = {x : i for i, x in enumerate(use)}
#     materials_map = {x : i for i, x in enumerate(materials)}
#     structtype_map = {x : i for i, x in enumerate(structtype)}
    
#     df["縣市"] = df["縣市"].map(city_map)
#     df["鄉鎮市區"] = df["鄉鎮市區"].map(town_map)
#     df["路名"] = df["路名"].map(road_map)
#     df["使用分區"] = df["使用分區"].map(district_map)
#     df["主要用途"] = df["主要用途"].map(use_map)
#     df["主要建材"] = df["主要建材"].map(materials_map)
#     df["建物型態"] = df["建物型態"].map(structtype_map)
    ################# preprocessing ######################
    df = df.drop(df[(df['土地面積']>10)].index)
    df = df.drop(df[(df['移轉層次']>=35)].index)
    df = df.drop(df[(df['總樓層數']>=45)].index)
    df = df.drop(df[(df['總樓層數']>=35)].index)
    df = df.drop(df[(df['建物面積']>=8)].index)
    df = df.drop(df[(df['車位面積']>=7)].index)
    df = df.drop(df[(df['橫坐標']<= 0)].index)
    df = df.drop(df[(df['主建物面積']>= 7)].index)
    df = df.drop(df[(df['陽台面積']>= 7)].index)
    df = df.drop(df[(df['附屬建物面積']>= 10)].index)
    df["使用分區"] = df["使用分區"].fillna("None")
    
    from sklearn.preprocessing import LabelEncoder
    cols = ('縣市', '鄉鎮市區', '路名', '使用分區', '主要用途', '主要建材', '建物型態')
    # process columns, apply LabelEncoder to categorical features
    for c in cols:
        lbl = LabelEncoder() 
        lbl.fit(list(df[c].values)) 
        df[c] = lbl.transform(list(df[c].values))
    
    df = zScoreCol(df, ['移轉層次', '總樓層數'])
    if mode == 'train':
        evaldata = df[['單價']]
        df = df.drop(['單價'], axis=1)
    ######################################################
    ############ use the external data ###################
    df = concat_externaldata(df, "./dataset/external_data/ATM資料.csv", 'ATM資料', radius=300.0)
    df = concat_externaldata(df, "./dataset/external_data/大學基本資料.csv", '大學基本資料', radius=3000.0)
    df = concat_externaldata(df, "./dataset/external_data/公車站點資料.csv", '公車站點資料', radius=300.0)
    df = concat_externaldata(df, "./dataset/external_data/火車站點資料.csv", '火車站點資料', radius=3000.0)
    df = concat_externaldata(df, "./dataset/external_data/金融機構基本資料.csv", '金融機構基本資料', radius=1000.0)
    df = concat_externaldata(df, "./dataset/external_data/便利商店.csv", '便利商店', radius=300.0)
    df = concat_externaldata(df, "./dataset/external_data/高中基本資料.csv", '高中基本資料', radius=3000.0)
    df = concat_externaldata(df, "./dataset/external_data/國小基本資料.csv", '國小基本資料', radius=3000.0)
    df = concat_externaldata(df, "./dataset/external_data/國中基本資料.csv", '國中基本資料', radius=3000.0)
    df = concat_externaldata(df, "./dataset/external_data/捷運站點資料.csv", '捷運站點資料', radius=300.0)
    df = concat_externaldata(df, "./dataset/external_data/郵局據點資料.csv", '郵局據點資料', radius=1000.0)
    df = concat_externaldata(df, "./dataset/external_data/腳踏車站點資料.csv", '腳踏車站點資料', radius=300.0)
    df = concat_externaldata(df, "./dataset/external_data/醫療機構基本資料.csv", '醫療機構基本資料', radius=3000.0)
#     df = zScoreCol(df, ['ATM資料', '大學基本資料', '公車站點資料', '火車站點資料',
#                        '金融機構基本資料', '便利商店', '高中基本資料', '國小基本資料', '國中基本資料',
#                        '捷運站點資料', '郵局據點資料', '腳踏車站點資料', '醫療機構基本資料'])
    
    df = normalizeCol(df, ['橫坐標', '縱坐標'])
    if mode == 'train':
        df = pd.concat([df, evaldata], axis=1)
    print(df)
    cols = df.columns
    cols_map = {x : i for i, x in enumerate(cols)}
#     df.columns = df.columns.map(cols_map)
#     print(cols_map)
    return df, cols_map
```
